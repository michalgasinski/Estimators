```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
s = np.random.exponential(scale=2,size=100)
x = np.linspace(0,6, 100)
plt.plot(x, stats.expon.pdf(x, 0, 1), label="Exponential distribution")
plt.show()
```

```python
means = []
medians = []
std = []
sample_sizes = range(30,10000)
for i in sample_sizes:
    s = np.random.exponential(scale=2,size=i)
    means.append(np.mean(s))  # Åšrednia
    medians.append(np.median(s))  # Mediana
    std.append(np.std(s))
    
```

```python
means_series = pd.Series(means)
medians_series = pd.Series(medians)
stds_series = pd.Series(std)
means_smooth = means_series.rolling(window=5).mean()
medians_smooth = medians_series.rolling(window=5).mean()
stds_smooth = stds_series.rolling(window=5).mean()
smoothed_sample_sizes = np.array(sample_sizes)[4:]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(smoothed_sample_sizes, means_smooth.dropna(), label='Smoothed Sample Mean', color='blue', marker='o', markersize=3)
plt.plot(smoothed_sample_sizes, medians_smooth.dropna(), label='Smoothed Sample Median', color='green', marker='s', markersize=3)
plt.plot(smoothed_sample_sizes, stds_smooth.dropna(), label='Smoothed Sample Std Dev', color='purple', marker='^', markersize=3)

# True Values
plt.axhline(y=2, color='red', linestyle='--', label='True Mean (Î¼=0)')
plt.axhline(y=np.log(2)/0.5, color='orange', linestyle='--', label='True Std Dev (Ïƒ=1)')

# Add titles and labels
plt.title('Convergence of Estimators with Smoothing (Window Size = 5)')
plt.xlabel('Sample Size (n)')
plt.ylabel('Estimator Value')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

We can observe that the estimators are unbiased (asymptotically unbiased in the case of the median), consistent, and efficient.
